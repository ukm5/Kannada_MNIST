# Standard NN

```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import OneHotEncoder

from keras.models import Sequential
from keras.layers import Dense
from keras.metrics import Accuracy
from keras.losses import CategoricalCrossentropy
```

```python
df_train = pd.read_csv('./train.csv')
df_train.head()
```

```python
y_train = df_train['label']
y_train.shape
```

```python
ohe = OneHotEncoder(dtype='int', sparse=False)
```

```python
y_train = ohe.fit_transform(y_train.to_numpy().reshape(-1,1))
```

```python
X_train = df_train.drop(columns=['label']).to_numpy()
```

```python
X_train.shape
```

```python
model = Sequential()
```

```python
model.add(Dense(units=256, activation='relu', input_shape = (784,)))
model.add(Dense(units=128, activation='relu'))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=32, activation='relu'))
model.add(Dense(units=10, activation='softmax'))
```

```python
model.compile('adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, batch_size=128, epochs=10, validation_split=0.10)
```
